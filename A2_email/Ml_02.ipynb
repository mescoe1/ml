```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("emails.csv")
```

**Checking Null Value**

```python
df.isnull().sum()
```

```python
null_count = df.isnull().sum()
#Filter column with value more than one null value
column_null = null_count[null_count >= 1]
column_null
```

```python
df.isnull().sum()
# all row having value as null
```

```python
# remove email column as it is not for use(i.e unnecessary)

x = df.iloc[:,1:3001]
```

```python
x.head()
```

```python
# output class

y = df.iloc[:,-1]
y.head()
```

**Training and Testing Model**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.2, random_state = 12)
```

```python
x_test
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

**K-Nearest Neighbour**

```python
# Fitting K-NN classifier to the training set
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn.fit(x_train, y_train)
```

**Prediction**

```python
y_pred = knn.predict(x_test)
y_pred
```

**Confusion Matric**

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
# Confusion matrix graph using heatmap

sns.heatmap(cm, annot=True, fmt=".2f", cmap= "Blues")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```

```python
accuracy_score(y_test, y_pred)
```

**Support Vector Machine**

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear', random_state=0)
svc.fit(x_train, y_train)
```

**Prediction**

```python
y_pred = svc.predict(x_test)
y_pred
```

**Confusion Matrix**

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
#confusion matrix graph representation

sns.heatmap(cm, annot=True, fmt = '.2f', cmap='Reds')
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```

**Accuracy Score**

```python
accuracy_score(y_test,y_pred)
```
