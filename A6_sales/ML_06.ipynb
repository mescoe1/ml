```python
X = df.iloc[: , [3,4]].values  #selects all rows and the columns at integer positions 3 and 4.
# iloc is a Pandas method used to select data from the DataFrame by integer-based location. It allows you to specify both the rows and columns you want to extract.
# .values converts the selected DataFrame subset into a NumPy array, which is often used as the input feature matrix for machine learning models.
```

**WCSS** is the sum of the squared distance between each point and the centroid in a cluster.

```python
# Initialize an empty list to store the within-cluster sum of squares (WCSS)
wcss = []
```

```python
# Determine the WCSS for a range of cluster numbers (e.g., 1 to 10)
for i in range(1,11):
  kmeans = KMeans(n_clusters= i, random_state=2)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
# Plot the WCSS values to identify the elbow point
plt.plot(range(1,11), wcss, 'bx-')
plt.title("Elbow Method")
plt.xlabel("Number of Cluster(Kvalue)")
plt.ylabel("WCSS")
plt.show()
```

```python
scale = StandardScaler()
scaled_data = scale.fit_transform(X)
```
